```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('year')
plt.ylabel('yield(tons per hectare)')
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.title('Crop Yields')
plt.xlabel('year')
plt.ylabel('yield(tone per hectare)')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.title('Crop Yields')
plt.xlabel('year')
plt.ylabel('yield(tone per hectare)')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples, marker='o', c='r')
plt.plot(years, oranges, marker='x', c='k', mew=2)

plt.title('Crop Yields')
plt.xlabel('year')
plt.ylabel('yield(tone per hectare)')
plt.legend(['Apples', 'Oranges'])
```

```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

```python
plt.figure(figsize=(13, 8))
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

```python
flowers_df=sns.load_dataset('iris')
```

```python
flowers_df.sepal_length
```

```python
plt.figure(figsize=(12 ,6))
plt.title("distribution of sepal length")
plt.hist(flowers_df.sepal_length)
```

```python
plt.hist(flowers_df.sepal_length, bins=5)
```

```python
import numpy as np
```

```python
np
```

```python
plt.hist(flowers_df.sepal_length, bins=np.arange(2, 5, 0.25))
```

```python
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25))
```

```python
plt.hist(flowers_df.sepal_width, bins=[1,3,4,5])
```

```python
setosa = flowers_df[flowers_df.species == 'setosa']
virginica = flowers_df[flowers_df.species == 'virginca']
versicolor = flowers_df[flowers_df.species == 'versicolor']
```

```python
plt.figure(figsize=(13, 7))
plt.hist([setosa.sepal_width, virginica.sepal_width, versicolor.sepal_width],
        bins=np.arange(2, 5, 0.25),
        stacked=True)

plt.legend(['setosa', 'virginica', 'versicolor'])
```
